```python
dicc_df = {'Costume' : list(dicc_disfraces.keys()),
        'Count' : list(dicc_disfraces.values())}
```

```python
df_categorias = pd.DataFrame(dicc_df)
```

```python
df_categorias.sort_values(by= 'Count', ascending = False, inplace = True)
```

```python
def agrupar_categorias(elemento):

    characters = ['Batman','Belle','Catwoman','Chucky','Harley Quinn','IT','Joker','Moana','Princess Jasmine','Queen of Hearts','Wonder Woman','Star Wars','The Wizard of Oz']
    animals = ['Chicken','Dinosaur','Lion','Mouse','Rabbit','Unicorn']
    scary = ['Monster','Skeleton','Scary Clown','Witch','Zombie']
    occupations = ['Cheerleader','Clown','Ninja','Pirate','Princess','Vikings','Superhero']

    if elemento in characters:
        return 'Characters and Movies'
    
    elif elemento in animals:
        return 'Animals'
    
    elif elemento in scary:
        return 'Scary Costumes'

    elif elemento in occupations:
        return 'Occupations'    
    
    else:
        return 'Others'
```

```python
df_categorias['Category'] = df_categorias['Costume'].apply(agrupar_categorias)
```
